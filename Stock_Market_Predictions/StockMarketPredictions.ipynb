```python
import yfinance as yf # Import the yahoo finance packages into the environment
```

```python
sp500 = yf.Ticker("^GSPC") # Brings through the data for the S&P 500
```

```python
sp500 = sp500.history(period="max") # Gets all the data of the S&P 500
```

```python
del sp500["Stock Splits"]
del sp500["Dividends"] # deltes the columns from the dataframe
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1) # Creates a tommorrow column which is equal to the close price of the day in front
```

```python
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
# Create a target column that will display a 0 if the tomorrow price is less than the close price 
# or a 1 if the tomorrow price is greater than the close
```

```python
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy() # This only takes lows where the index is at least 1990 january the 1st
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier
# Using random forest as it trains mutiple decsion trees anc averaging the results. This stops overfitting and non-linear relationships.
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)
# Creates 100 different decision trees, Helps protect against overfitting, Random state means same results when the model is run mutiple times
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]
# Splits the data into train and test set. The test set is the forst 100 rows.
predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
# Trains model using the columns as predictors and guesses the Traget column value.
```

```python
from sklearn.metrics import precision_score 
# This checks if the market actully went up when predicted it would
preds = model.predict(test[predictors])
#This tests the model using our test set
```

```python
import pandas as pd 
# imports Pandas
preds = pd.Series(preds, index=test.index) 
# Turn the predictors into a pandas series
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"]) # Fitting the model using the training predictors
    preds = model.predict(test[predictors]) # Generating the predictors
    preds = pd.Series(preds, index=test.index, name="Predictions") #Turn it into a series
    combined = pd.concat([test["Target"], preds], axis=1) # Combining everything together
    return combined # Returning the Combined dataframe
```

```python
def backtest(data, model, predictors, start=2500, step=250): # Start takes 10 years of data and trains the model. While step trains the next year.
    all_predictions = [] # Creates a list of all dataframes. A dataframe is a single year

    for i in range(start, data.shape[0], step): #Cretaes a funtion to loop acroos the data year by year and make predictions for all years apart from the first 10.
        train = data.iloc[0:i].copy() 
        test = data.iloc[i:(i+step)].copy() # Splits the train and test data
        predictions = predict(train, test, predictors, model)  # Generates the predicitons
        all_predictions.append(predictions) # append the predicitons for that year
    return pd.concat(all_predictions) # combines all dataframe into one
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts() # Counts how many times the model preicted the market would go up or down
```

```python
precision_score(predictions["Target"], predictions["Predictions"]) # How accurate is this model
```

```python
predictions["Target"].value_counts() / predictions.shape[0] # This count the number of preictions and divides by number of rows total
```

```python
horizons = [2,5,60,250,1000] # Defines a list of time horizons
new_predictors = [] # Initilize an empty list of newly created predictor columns

for horizon in horizons: # Loop through each horizon
    rolling_averages = sp500.rolling(horizon).mean() # Calculate the rolling average for each horizon time frame

    ratio_column = f"Close_Ratio_{horizon}" # Create a new column for the ratio of current price to its rolling average
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"] # Compute the ration and add new column to the data frame

    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"] # Rolling sum of the 'Target' Column, shifted by 1 day to avoid bias
    
    new_predictors += [ratio_column, trend_column] # Add new columns to the predictors
```

```python
sp500 = sp500.dropna() # gets rid of any columns with NA
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"]) # Fitting the model using the training predictors
    preds = model.predict_proba(test[predictors])[:,1] # Generating the predictors in probability
    preds[preds >= .6] = 1
    preds [preds < .6] = 0 # Makes so value in the prediction column to only display 1 or 0 if the probability is more than 0.6
    preds = pd.Series(preds, index=test.index, name="Predictions") #Tunr it into a series
    combined = pd.concat([test["Target"], preds], axis=1) # Combining everything together
    return combined # Returning the Combined dataframe
```

```python
predictions = backtest(sp500, model, new_predictors)
```
